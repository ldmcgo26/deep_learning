```python
# Training set
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

dataset_dir = "/content/VisDrone/VisDrone2019-DET-train"
image_dir = os.path.join(dataset_dir, "images")
ann_dir = os.path.join(dataset_dir, "annotations")

image_files = sorted(os.listdir(image_dir))[:1000]

images_np = []
annotations_np = []

for img_file in image_files:

    img_path = os.path.join(image_dir, img_file)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    ann_file = img_file.replace('.jpg', '.txt')
    ann_path = os.path.join(ann_dir, ann_file)

    boxes = []
    classes = []

    with open(ann_path, 'r') as f:
        for line in f:
            x, y, w, h, score, obj_class, trunc, occ = map(int, line.strip().split(','))
            boxes.append([x, y, x + w, y + h])
            classes.append(obj_class)

    images_np.append(img)
    annotations_np.append({
        'boxes': np.array(boxes, dtype=np.float32),
        'labels': np.array(classes, dtype=np.int32)
    })

print("Loaded {} images.".format(len(images_np)))

plt.imshow(images_np[0])
plt.axis('off')
plt.show()

print("Boxes:\n", annotations_np[0]['boxes'])
print("Labels:\n", annotations_np[0]['labels'])
```

```python
"""""""""""""""""""""""""""""""""

Description:

The remainder of the file includes various approaches to exploration of the
data. While not strictly necessary to the completion of our project and final
preparation of the datasets for model training, the goal is simply to gain a
greater familiarity with - and understanding of - the data we will be working
with. The data exploration found in the below blocks includes:

1) A method for taking in an image and its corresponding annotations (both in
   numpy array format) to display a sample with drawn-in bounding boxes
   according to the VisDrone labels map.
2) A method for iterating through all images in a representative subset to
   count and remove all instances of a dedicated label 0: "Ignored" as defined
   by VisDrone image labeling.
3) A method for iterating through all images in a representative subset to tally
   the total number of each label occurring across the subset. It loops through
   a complex list of annotations and prints out the label IDs and their
   corresponding counts, then accumulates these individual counts into a master
   list before plotting a visual representation.
4) Finally, a method for iterating through all images in a representative subset
   to determine the image with the least identified annotations / bounding boxes
   and the image with the most identified annotations / bounding boxes.


"""""""""""""""""""""""""""""""""
```

```python
# Function to construct and display image with corresponding drawn-in labels
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def display_sample(image, boxes, labels=None, label_map=None):

    fig, ax = plt.subplots(1, figsize=(12, 12))

    ax.imshow(image)

    for i, box in enumerate(boxes):
        xmin, ymin, xmax, ymax = box

        rect = patches.Rectangle(
            (xmin, ymin),
            xmax - xmin,
            ymax - ymin,
            linewidth=2,
            edgecolor='red',
            facecolor='none'
        )

        ax.add_patch(rect)

        if labels is not None:
            label_id = labels[i]
            label_text = str(label_id)
            if label_map and label_id in label_map:
                label_text = label_map[label_id]

            ax.text(
                xmin,
                ymin - 5,
                label_text,
                fontsize=12,
                color='yellow',
                bbox=dict(facecolor='red', alpha=0.5)
            )

    plt.axis('off')
    plt.show()
```

```python
# Count and remove instances of "Ignored" label
total_ignored_annotations = 0

for i, ann in enumerate(annotations_np):
    boxes = ann['boxes']
    labels = ann['labels']

    valid_indices = labels != 0

    num_ignored = np.sum(labels == 0)
    total_ignored_annotations += num_ignored

    annotations_np[i]['boxes'] = boxes[valid_indices]
    annotations_np[i]['labels'] = labels[valid_indices]

print(f"Total ignored annotations removed: {total_ignored_annotations}")
```

```python
# Is this necessary here?

import torch
from torch.utils.data import Dataset
import torchvision.transforms.functional as F
import numpy as np
import cv2

class VisDroneDataset(Dataset):
    def __init__(self, images_np, annotations_np, resize_to=(640, 640), transforms=None, device='cpu'):
        """
        Args:
            images_np (list): List of numpy arrays representing images.
            annotations_np (list): List of dicts with 'boxes' and 'labels'.
            resize_to (tuple): Desired image size (H, W) after resizing.
            transforms (callable, optional): Additional torchvision or albumentations transforms.
            device (str): 'cpu' or 'cuda' depending on where you want your data.
        """
        self.images = images_np
        self.annotations = annotations_np
        self.resize_to = resize_to  # (H, W)
        self.transforms = transforms
        self.device = device

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        # Get image and annotation
        image = self.images[idx]  # shape: (H, W, 3)
        ann = self.annotations[idx]

        # Original image size
        orig_h, orig_w = image.shape[:2]

        # Resize image
        resized_image = cv2.resize(image, (self.resize_to[1], self.resize_to[0]))  # (W, H)

        # Convert to tensor and normalize [0, 1]
        resized_image = torch.from_numpy(resized_image).permute(2, 0, 1).float() / 255.0  # (3, H, W)

        # Scale bounding boxes according to resize
        scale_x = self.resize_to[1] / orig_w
        scale_y = self.resize_to[0] / orig_h

        boxes = ann['boxes'].copy()
        boxes[:, [0, 2]] = boxes[:, [0, 2]] * scale_x  # Scale x coords
        boxes[:, [1, 3]] = boxes[:, [1, 3]] * scale_y  # Scale y coords

        # Labels
        labels = ann['labels'].copy()

        # Convert boxes and labels to tensors
        boxes = torch.as_tensor(boxes, dtype=torch.float32)
        labels = torch.as_tensor(labels, dtype=torch.int64)

        # Create target dictionary (common convention for PyTorch detectors)
        target = {}
        target['boxes'] = boxes
        target['labels'] = labels
        target['image_id'] = torch.tensor([idx])

        # Apply additional transforms (optional)
        if self.transforms:
            resized_image, target = self.transforms(resized_image, target)

        return resized_image.to(self.device), target
```

```python
# Function to compute count of each label in a given dict of image annotations
import numpy as np

def count_labels_in_image(annotations):

    label_counts = {}

    for label in annotations['labels']:
        if label != 0:
            if label not in label_counts:
                label_counts[label] = 1
            else:
                label_counts[label] += 1

    return dict(sorted(label_counts.items()))


sample_index = 0
sample_annotations = annotations_np[sample_index]

label_counts = count_labels_in_image(sample_annotations)
print("Label counts for the image:", label_counts)
```

```python
# Function to iterate through all images of a dataset and accumulate individual
# label counts into single master list of label counts
from collections import defaultdict

def accumulate_label_counts(dataset_annotations):

    master_label_counts = defaultdict(int)

    num_images = len(dataset_annotations)
    print(f"Number of images in the dataset: {num_images}")

    for annotations in dataset_annotations:
        label_counts = count_labels_in_image(annotations)
        for label, count in label_counts.items():
            master_label_counts[label] += count

    return dict(sorted(master_label_counts.items()))

master_counts = accumulate_label_counts(annotations_np)

print("Master label counts across the entire dataset:", master_counts)
```

```python
# Using master list of label counts, display frequency of labels in a chart
import matplotlib.pyplot as plt

def plot_label_counts(master_label_counts):

    labels = list(master_label_counts.keys())
    counts = list(master_label_counts.values())

    plt.figure(figsize=(10, 6))
    plt.bar(labels, counts, color='skyblue')
    plt.title("Label Counts in the Dataset", fontsize=14)
    plt.xlabel("Label ID", fontsize=12)
    plt.ylabel("Count", fontsize=12)

    qualitative_labels = ['Pedestrian', 'Person', 'Car', 'Van', 'Bus', 'Truck', 'Motor', 'Bicycle', 'Awning-tricycle', 'Tricycle', 'Other']
    plt.xticks(labels, qualitative_labels, rotation=90, ha='right')
    plt.show()

plot_label_counts(master_counts)
```
